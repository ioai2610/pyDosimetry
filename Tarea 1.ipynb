# Problema 1

Sea $\vec{x}=(x_{1},x_{2})\in\mathbb{R}^{2}$. Considere las siguientes definiciones de norma en $\mathbb{R}^{2}$.   
* $\|\vec{x}\|_{1}=\lvert\ x_{1}\rvert+\lvert\ x_{2}\rvert$.
* $\|\vec{x}\|_{p}=\left(\lvert\ x_{1}\rvert^{p}+\lvert\ x_{2}\rvert^{p}\right)^{1/p}$, con $p>0$.
* $\|\vec{x}\|_{\infty}=\text{sup}\lbrace{\lvert\ x_{1}\rvert,\lvert\ x_{2}\rvert\rbrace}$.  

Realice un programa para graficar en $\mathbb{R}^{2}$ las bolas abiertas siguientes:

(1) $\vec{B_{1}}(0,1)=\lbrace{\vec{x}\in\mathbb{R}^{2}:\|\vec{x}\|_{1}<1\rbrace}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el centro y radio de la bola abierta
centro = [0,0]
radio = 1

# Generamos una malla de puntos en el cuadrado [-2, 2] x [-2, 2]
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

# Uso de meshgrid
# https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html#numpy.meshgrid
xx, yy = np.meshgrid(x, y)

# Calculamos la norma L1 de cada punto de la malla e imponemos la condición de centrado
norma1 = np.abs(xx-centro[0]) + np.abs(yy-centro[1])

# Graficamos los puntos dentro de la bola abierta
# Uso de contour
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html
plt.contour(xx, yy, (norma1 < radio))

# Establecemos los límites de los ejes y mostrar el gráfico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title("Bola abierta B(0,1) con norma L 1")
plt.xlabel("Irving Orlando Ayala Iturbe")
plt.show()
```

# Problema 3

Sea $\textit{H}$ un operador lineal del espacio euclideano $\mathbb{E}^{N}$ a $\mathbb{E}^{M}$.  
Considere que cada
entrada es arbitraria, es decir, $H_{ij}\in\mathbb{E}$. Realice una rutina numérica para determinar el operador
adjunto $\textit{H}$.  
(Puede hacerlo para valores de $\textit{N}$ y $\textit{M}$ fijos, digamos $N = 2$ y $M = 3$, y entradas
fijas o bien intente hacerlo en general teniendo como información de entrada el valor de $N$ y
$M$).

```python
# funcion p/calcular traspuesta de una matriz asociada a un operador
# solicitamos las entradas en forma de lista de columnas de listas de renglones, en la forma:
# [[entrada 1-1,entrada 1-2, ..., entrada 1-n],...,[entrada n-1,entrada n-2, ..., entrada n-n]]
def transOp(matrix): 
    print('La matriz ingresada es: ')
    for i in matrix:
        print(i)
    print('La matriz transpuesta conjugada es: ')
    # reordenamos en una sola lista de tal manera que se cumpla la condicion de transpuesta
    # en cada entrada "matrix[j][i]" usamos el metodo ".conjugate()" para congujar una entrada compleja
    transp = [[matrix[j][i].conjugate() for j in range(len(matrix))] for i in range(len(matrix[0]))]
    for e in transp:
        print(e)
```

Ejemplo de uso

```python
kk = complex(1,2)
uu = complex(3,4)
yy = complex(5,6)
l = [[1,2,3],[kk,uu,yy]]
transOp(l)
```
